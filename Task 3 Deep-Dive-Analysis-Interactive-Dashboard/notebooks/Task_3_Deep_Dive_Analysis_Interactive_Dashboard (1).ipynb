```python
import pandas as pd

df = pd.read_excel("/content/Online_Retail (1).xlsx")

df = df.dropna(subset=['CustomerID'])
df = df[df['Quantity'] > 0]
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['Revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')
```

```python
df['CohortMonth'] = df.groupby('CustomerID')['InvoiceDate'] \
                       .transform('min') \
                       .dt.to_period('M')
```

```python
df['CohortIndex'] = (df['InvoiceMonth'].astype(int) -
                     df['CohortMonth'].astype(int))
```

```python
cohort_data = df.groupby(['CohortMonth', 'CohortIndex']) \
                ['CustomerID'].nunique().reset_index()

cohort_counts = cohort_data.pivot(index='CohortMonth',
                                   columns='CohortIndex',
                                   values='CustomerID')
```

```python
cohort_sizes = cohort_counts.iloc[:, 0]

retention = cohort_counts.divide(cohort_sizes, axis=0)

retention
```
