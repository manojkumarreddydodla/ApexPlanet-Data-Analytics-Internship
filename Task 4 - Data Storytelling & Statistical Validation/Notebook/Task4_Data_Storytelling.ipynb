```
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_excel("/content/Online_Retail (1).xlsx")

# Basic cleaning
df.dropna(inplace=True)
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]

# Create Total Amount
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']

# Create customer level dataframe
customer_df = df.groupby('CustomerID').agg({
    'TotalAmount': 'sum',
    'InvoiceNo': 'nunique'
}).reset_index()

customer_df.columns = ['CustomerID', 'TotalSpend', 'TotalOrders']

customer_df.head()
```

```
from scipy.stats import ttest_ind

repeat_customers = customer_df[customer_df['TotalOrders'] > 1]['TotalSpend']
one_time_customers = customer_df[customer_df['TotalOrders'] == 1]['TotalSpend']

t_stat, p_value = ttest_ind(repeat_customers, one_time_customers)

print("T Statistic:", t_stat)
print("P Value:", p_value)
```
